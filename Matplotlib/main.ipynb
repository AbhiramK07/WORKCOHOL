```python
fig, ax = plt.subplots(figsize = (8,4))
x1 = np.linspace(0,100,20)
x2 = np.linspace(0,100,20)
x3 = np.linspace(0,100,20)

y1 = np.sin(x1)
y2 = np.cos(x2)
y3 = np.tan(x3)

ax.plot(x1,y1, label = "Sin")
ax.plot(x2,y2, label = "Cos")
ax.plot(x3,y3, label = "Tan")

ax.legend(loc='best', framealpha = 0.5, prop = {'size': 'small'})


```

```python
fig.savefig("Chart.png",dpi = 125)
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(nrows=3, ncols=1, figsize = (8,4))
x1 = np.linspace(0,100,20)
x2 = np.linspace(0,100,20)
x3 = np.linspace(0,100,20)

y1 = np.sin(x1)
y2 = np.cos(x2)
y3 = np.tan(x3)

ax1.plot(x1,y1, label = "Sin")
ax2.plot(x2,y2, label = "Cos")
ax3.plot(x3,y3, label = "Tan")

ax1.legend(loc='best', framealpha = 0.5, prop = {'size': 'small'})
ax2.legend(loc='best', framealpha = 0.5, prop = {'size': 'small'})
ax3.legend(loc='best', framealpha = 0.5, prop = {'size': 'small'})

ax1.grid(True)
ax2.grid(True)
ax3.grid(True)


```

```python
data = np.array([15, 12, 9])
labels = ['Python', 'Javascript', 'Java']

colors = ["orange", "cyan", "tan"]      #colors

explode = (0,0,0.1)      #separate pie
fig,ax = plt.subplots(figsize=(8,4))
ax.pie(data, labels = labels, colors=colors, explode=explode, autopct='%1.1f%%')

ax.axis('equal')            # single area
fig.suptitle("Marks")       # Title


```

```python
np.random.seed(3)
x = 4+np.random.normal(0,2,24)
y = 4+np.random.normal(0,2,len(x))


```

```python
size = np.random.uniform(15, 80, len(x))

colors = np.random.uniform(15, 80, len(x))
```

```python
fig, ax = plt.subplots()

ax.scatter(x,y, s = size, c = colors, vmin = 0, vmax = 100)
ax.set(xlim=(0,10),xticks=np.arange(1,10), ylim=(0,10), yticks = np.arange(1,10))
plt.show()
```

```python
np.random.seed(1)

x = np.linspace(0,8,16)
y1 = 3+4*x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1+2*x/8 + np.random.uniform(0.0, 0.5, len(x))



```

```python
fig, ax = plt.subplots()

ax.fill_between(x,y1,y2,alpha = 0.5)
ax.plot(x, (y1+y2)/2)
plt.show()
```
